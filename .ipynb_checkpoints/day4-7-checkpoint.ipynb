## Day 4 (Rock paper scissors game)

```python
import random
import numpy as np
from IPython.display import clear_output
import time


random.seed(688)

rock = """
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
"""

paper = """
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
"""

scissors = """
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
"""
def choice_print(choice):
    if choice == 0:
        print(rock)
    elif choice == 1:
        print(paper)
    else:
        print(scissors)
                
def losowanie():
    my_choice = int(input())
    while my_choice not in (0,1,2):
        my_choice = int(input())
    computer_choice = random.choice([0,1,2])
    
    return my_choice, computer_choice

def rematch():
    print('Rematch.')
    time.sleep(1)
    clear_output()
    print('Type again.\n')
    gra()

def play_again():
    time.sleep(1)
    clear_output()
    print('What do you choose? Type 0 for Rock, 1 for Paper or  2 for Scissors.\n')
    gra()

def gra():
    my_choice, computer_choice = losowanie()
    choice_print(my_choice)
    print('Computer choice:\n')
    choice_print(computer_choice)
    sub = my_choice - computer_choice
    if sub == 0:
        rematch()
    elif sub in (-2, 1):
        print('You Win!')
        want_again()
    else:
        print('You Lose')
        want_again()

def want_again():
    question = input('Want to play again? (Y/N)\n')
    
    if question == 'Y':
        play_again()
    elif question == 'N':
        print('End of Game :\ )')
    else:
        print('You are an analfabeth !')
```

```python
print('What do you choose? Type 0 for Rock, 1 for Paper or  2 for Scissors.\n')
gra()
```

## Day 5

```python
import string
import random

all_l = list(string.ascii_letters)
all_n = list(string.digits)
all_s = list(string.punctuation)

print('Welcome to the PyPassword Generator')
l = int(input('How many letters would you like in your password?'))
s = int(input('How many symbols would you like?'))
n = int(input('How many numbers would you like?'))

def how_many(my_list, num):
    return random.choices(my_list, k=num)

choice_l = how_many(all_l, l)
choice_s = how_many(all_s, s)
choice_n = how_many(all_n, n)

stack = choice_l + choice_n + choice_s
random.shuffle(stack)
my_password = "".join(stack)
print(stack)
print(f'Here is your password: {my_password}')
```

## Day 7 (Hangman game)

```python
HANGMANLOGO= '''
 _                                             
| |                                            
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __  
| '_ \ / _` | '_ \ / _` | '_ ` _ \ / _` | '_ \ 
| | | | (_| | | | | (_| | | | | | | (_| | | | |
|_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                    __/ |                      
                   |___/  
'''

HANGMANPICS = ['''
  +---+
  |   |
      |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========''', '''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========''']
fireworks = '''
               *    *
   *         '       *       .  *   '     .           * *
                                                               '
       *                *'          *          *        '
   .           *               |               /
               '.         |    |      '       |   '     *
                 \*        \   \             /
       '          \     '* |    |  *        |*                *  *
            *      `.       \   |     *     /    *      '
  .                  \      |   \          /               *
     *'  *     '      \      \   '.       |
        -._            `                  /         *
  ' '      ``._   *                           '          .      '
   *           *\*          * .   .      *
*  '        *    `-._                       .         _..:='        *
             .  '      *       *    *   .       _.:--'
          *           .     .     *         .-'         *
   .               '             . '   *           *         .
  *       ___.-=--..-._     *                '               '
                                  *       *
                *        _.'  .'       `.        '  *             *
     *              *_.-'   .'            `.               *
                   .'                       `._             *  '
   '       '                        .       .  `.     .
       .                      *                  `
               *        '             '                          .
     .                          *        .           *  *
             *        .                                    '
'''
HANGMANPICS = HANGMANPICS[::-1]
#Word bank of animals
words = ('ant baboon badger bat bear beaver camel cat clam cobra cougar '
         'coyote crow deer dog donkey duck eagle ferret fox frog goat '
         'goose hawk lion lizard llama mole monkey moose mouse mule newt '
         'otter owl panda parrot pigeon python rabbit ram rat raven '
         'rhino salmon seal shark sheep skunk sloth snake spider '
         'stork swan tiger toad trout turkey turtle weasel whale wolf '
         'wombat zebra ').split()
```

```python
import random
from IPython.display import clear_output

game_word = random.choice(words)

progres_list = list('_' * len(game_word))
password_list = list(game_word)
toolbar = ' '.join(progres_list)
counter = 6
end_of_game = False

print(HANGMANLOGO)
print(toolbar)
print(HANGMANPICS[counter])

while not end_of_game:
    
    guess = input('Guess ad letter:')
    clear_output()
    
    if guess == "happynewyear":
        print(fireworks)
        print('Happy New Year !!!')
        break
    
    if guess in progres_list:
        print(HANGMANLOGO)
        print(toolbar)
        print(HANGMANPICS[counter])
        print(f'You already chose guessed {guess}')
        
    elif guess in password_list:
        my_index = password_list.index(guess)
        progres_list[my_index] = guess
        password_list[my_index] = '_'
        toolbar = ' '.join(progres_list)
        print(HANGMANLOGO)
        print(toolbar)
        print(HANGMANPICS[counter])
        
    elif guess not in password_list:
        counter -=1
        print(HANGMANLOGO)
        print(toolbar)
        print(HANGMANPICS[counter])
        print(f"You guessed {guess}, that's not in the word. You lose a life")

        if counter == 0:
            end_of_game = True
            print('You lose!')
 
    if '_' not in progres_list:
        end_of_game = True
        print("You win!")
```
