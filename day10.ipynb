```python
from tkinter import *
root = Tk()
root.title('Simple calculator')

entry = Entry(root, width=36, borderwidth=5)
entry.grid(row=0, column=0 , columnspan=4, padx=5, pady=5)

def click(number):
    temp = entry.get()
    entry.delete(0, END)
    entry.insert(0 , str(temp) + str(number))
    
def click_dot():
    temp = entry.get()
    entry.delete(0, END)
    entry.insert(0 , str(temp) + ".")
    
def clear():
    entry.delete(0, END)

def add():
    first = entry.get()
    global f_number
    global relation
    relation = 'addition'
    f_number = float(first)
    entry.delete(0, END)

def subtract():
    first = entry.get()
    global f_number
    global relation
    relation = 'subtraction'
    f_number = float(first)
    entry.delete(0, END)
    
def multiply():
    first = entry.get()
    global f_number
    global relation
    relation = 'multiplication'
    f_number = float(first)
    entry.delete(0, END)
    
def divide():
    first = entry.get()
    global f_number
    global relation
    relation = 'division'
    f_number = float(first)
    entry.delete(0, END)
    
def percentage():
    first = entry.get()
    global f_number
    f_number = float(first)
    entry.delete(0, END)   
    entry.insert(0, round(f_number /100,4))
        
def equal(): 
    second = entry.get()
    entry.delete(0, END)
    
    if relation == 'addition':
        entry.insert(0, round(f_number + float(second), 4)) 
    if relation == 'subtraction':
        entry.insert(0, round(f_number - float(second), 4))   
    if relation == 'multiplication':
        entry.insert(0, round(f_number * float(second), 4))   
    if relation == 'division':
        entry.insert(0, round(f_number / float(second), 4))   
   
    

button0 = Button(root, text='0', padx=50, pady=16, command=lambda: click(0))
button1 = Button(root, text='1', padx=20, pady=16, command=lambda: click(1))
button2 = Button(root, text='2', padx=20, pady=16, command=lambda: click(2))
button3 = Button(root, text='3', padx=20, pady=16, command=lambda: click(3))
button4 = Button(root, text='4', padx=20, pady=16, command=lambda: click(4))
button5 = Button(root, text='5', padx=20, pady=16, command=lambda: click(5))
button6 = Button(root, text='6', padx=20, pady=16, command=lambda: click(6))
button7 = Button(root, text='7', padx=20, pady=16, command=lambda: click(7))
button8 = Button(root, text='8', padx=20, pady=16, command=lambda: click(8))
button9 = Button(root, text='9', padx=20, pady=16, command=lambda: click(9))

button_mult = Button(root, text='*', padx=20, pady=16, command=multiply)
button_div = Button(root, text='/', padx=20, pady=16, command=divide)
button_sub = Button(root, text='-', padx=20, pady=16, command=subtract)
button_add = Button(root, text='+', padx=18.5, pady=16, command=add)

button_AC = Button(root, text='AC', padx=44, pady=16, command=clear)
button_dot = Button(root, text='.', padx=22, pady=16, command=click_dot)
button_perc = Button(root, text='%', padx=18.5, pady=16, command=percentage)
button_equal = Button(root, text='=', padx=18.5, pady=16, command=equal)

button_AC.grid(row=1, column=0, columnspan=2)
button_dot.grid(row=5, column=2)
button_perc.grid(row=1, column=2)
button_div.grid(row=1, column=3)

button0.grid(row=5, column=0, columnspan=2)
button1.grid(row=4, column=0)
button2.grid(row=4, column=1)
button3.grid(row=4, column=2)
button4.grid(row=3, column=0)
button5.grid(row=3, column=1)
button6.grid(row=3, column=2)
button7.grid(row=2, column=0)
button8.grid(row=2, column=1)
button9.grid(row=2, column=2)

button_mult.grid(row=2, column=3)
button_sub.grid(row=3, column=3)
button_add.grid(row=4, column=3)
button_equal.grid(row=5, column=3)


root.mainloop()
```
